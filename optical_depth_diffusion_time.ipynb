```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.interpolate import interp1d
```

```python
log_csm_rho_data = np.log10(np.loadtxt('CSM_density/SN2023ixf.txt').T) # [0] r (cm); [1] density (g/cm^3)
log_csm_opacity_data = np.log10(np.loadtxt('CSM_tau/SN2023ixf.txt').T) # [0] opacity (cm^2/g) ; [1] density (g/cm^3)
```

```python
log_csm_rho_interp = interp1d(log_csm_rho_data[0],log_csm_rho_data[1],kind='linear',bounds_error=False,fill_value=np.nan)
log_csm_opacity_interp = interp1d(log_csm_rho_data[1],log_csm_rho_data[0],kind='linear',bounds_error=False,fill_value=np.nan)
```

```python
def csm_rho(r):
    r = np.log10(r)
    return 10**log_csm_rho_interp(r)

def csm_opacity(rho):
    rho = np.log10(rho)
    return 10**log_csm_opacity_interp(rho)
```
